```python
import numpy as np
import pandas as pd
import re
import pandas as pd
```

```python
f=open('./data/smallreldistribution.txt', encoding="utf-8")

data = f.readlines()  # 直接将文件中按行读到list里，效果与方法2一样

f.close()  # 关
print(data) #返回list
```

```python
data
```

```python
i=0
cat = {}
for line in data:
    elements = re.split(r'[-\s./,;!?]+', line)[:2]
    if elements[0] not in cat.keys():
        temp = []
        temp.append(elements[1])
        cat[elements[0]] = temp
    else:
        cat[elements[0]].append(elements[1])
cat
```

```python
# 利用pandas模块先建立DateFrame类型，然后将两个上面的list存进去
result_excel = pd.DataFrame()
result_excel["二级指标"] = cat.keys()
result_excel["三级指标"] = cat.values()
# 写入excel
result_excel.to_excel('./data/labels.xls')
```
